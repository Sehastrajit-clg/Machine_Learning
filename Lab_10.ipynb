```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

# Load the dataset (flame.csv or toy.csv)
data = pd.read_csv("flame.csv")  # Replace "flame.csv" with "toy.csv" for the other dataset

# Extract the features from the dataset
X = data.iloc[:, :-1].values

# Perform K-means clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Plotting the original clusters
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=data["cluster"], cmap="viridis")
plt.title("Original Clusters")

# Plotting the clusters obtained by K-means
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, color="red")
plt.title("Clusters Obtained by K-means")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset (flame.csv or toy.csv)
data = pd.read_csv("flame.csv")  # Replace "flame.csv" with "toy.csv" for the other dataset

# Extract the features from the dataset
X = data.iloc[:, :-1].values

# Perform K-means clustering with different parameters
n_clusters = 2  # Change the number of clusters as desired

# Example 1: Euclidean distance, default parameters
kmeans_default = KMeans(n_clusters=n_clusters, random_state=0)
kmeans_default.fit(X)
labels_default = kmeans_default.labels_
silhouette_default = silhouette_score(X, labels_default)

# Example 2: Manhattan distance, increase max_iter
kmeans_manhattan = KMeans(n_clusters=n_clusters, random_state=0, max_iter=100)
kmeans_manhattan.fit(X)
labels_manhattan = kmeans_manhattan.labels_
silhouette_manhattan = silhouette_score(X, labels_manhattan)

# Plotting the clusters obtained with different parameters
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_default, cmap="viridis")
plt.title("Default Parameters (Silhouette: {:.2f})".format(silhouette_default))
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels_manhattan, cmap="viridis")
plt.title("Manhattan Distance, max_iter=100 (Silhouette: {:.2f})".format(silhouette_manhattan))

# Show the plot
plt.tight_layout()
plt.show()

```
